# Part 3: One Variable Linear Regression (30 points)

In this part, we'll perform some one-variable linear regression analysis on the world happiness data.

&#9989;  **Question 3.1 (6 points)**: Using the `OLS` method in `statsmodels`, perform a linear regression to predict the `happiness_score` using the `healthy_life_expectancy` column separately for 2022 and 2023 and **display the results summaries** for both years. Remember that you may need to use the `add_constant()` method to make sure `OLS` fits a general line $y = ax+b$ to the data instead of a line through the origin $y = ax$. 

````python
# Put your code here

import statsmodels.api as sm

# For 2022
X_2022 = sm.add_constant(df_2022['healthy_life_expectancy'])
y_2022 = df_2022['happiness_score']

model_2022 = sm.OLS(y_2022, X_2022).fit()
print("Linear Regression Results for 2022:")
print(model_2022.summary())

# For 2023
X_2023 = sm.add_constant(df_2023['healthy_life_expectancy'])
y_2023 = df_2023['happiness_score']

model_2023 = sm.OLS(y_2023, X_2023).fit()
print("\nLinear Regression Results for 2023:")
print(model_2023.summary())




# For 2022
intercept_2022 = model_2022.params['const']
coefficient_2022 = model_2022.params['healthy_life_expectancy']
print("Best fit equation for 2022: happiness_score = {:.2f} + {:.2f} * healthy_life_expectancy".format(intercept_2022, coefficient_2022))

# For 2023
intercept_2023 = model_2023.params['const']
coefficient_2023 = model_2023.params['healthy_life_expectancy']
print("Best fit equation for 2023: happiness_score = {:.2f} + {:.2f} * healthy_life_expectancy".format(intercept_2023, coefficient_2023))

````

&#9989;  **Question 3.3 (6 points)**: Now make scatter plots of `happiness_score` (x-axis) vs. `healthy_life_expectancy` (on y-axis) for 2022 and 2023 on the same plot. Use different colors for the different years. Also plot the best fit lines on the same plot with same colors as the corresponding scatter plots. Label the axes, and add a legend, and give the plot a title.

````python
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm 

# Scatter plot for 2022
plt.scatter(df_2022['happiness_score'], df_2022['healthy_life_expectancy'], color='blue', label='2022')

# Scatter plot for 2023
plt.scatter(df_2023['happiness_score'], df_2023['healthy_life_expectancy'], color='red', label='2023')

# Best fit line for 2022
x_2022 = np.linspace(df_2022['happiness_score'].min(), df_2022['happiness_score'].max(), 100)
X_2022 = sm.add_constant(x_2022)
model_2022 = sm.OLS(df_2022['healthy_life_expectancy'], sm.add_constant(df_2022['happiness_score'])).fit()  # Fit linear regression model
y_2022 = model_2022.predict(X_2022)
plt.plot(x_2022, y_2022, color='blue', linestyle='--', label='Best Fit Line 2022')

# Best fit line for 2023
x_2023 = np.linspace(df_2023['happiness_score'].min(), df_2023['happiness_score'].max(), 100)
X_2023 = sm.add_constant(x_2023)
model_2023 = sm.OLS(df_2023['healthy_life_expectancy'], sm.add_constant(df_2023['happiness_score'])).fit()  # Fit linear regression model
y_2023 = model_2023.predict(X_2023)
plt.plot(x_2023, y_2023, color='red', linestyle='--', label='Best Fit Line 2023')

# Labeling axes and adding title
plt.xlabel('Happiness Score')
plt.ylabel('Healthy Life Expectancy')
plt.title('Scatter Plot of Happiness Score vs. Healthy Life Expectancy (2022-2023)')

# Adding legend
plt.legend()

# Display plot
plt.grid(True)
plt.show()
````

**&#9989; Do this:** **Question 3.5 (4 points)**: Use `plot_regress_exog` to investigate the distribution of residuals in your model fits. Make sure to create a large enough figure so that everything is easily visible. Make separate plots for 2022 and 2023.

````python
import statsmodels.api as sm

# Fit linear regression models for 2022 and 2023
model_2022 = sm.OLS(df_2022['healthy_life_expectancy'], sm.add_constant(df_2022['happiness_score'])).fit()
model_2023 = sm.OLS(df_2023['healthy_life_expectancy'], sm.add_constant(df_2023['happiness_score'])).fit()

# Plot residuals for 2022
plt.figure(figsize=(10, 6))
sm.graphics.plot_regress_exog(model_2022, 'happiness_score')
plt.title('Residuals Distribution for 2022')

# Plot residuals for 2023
plt.figure(figsize=(10, 6))
sm.graphics.plot_regress_exog(model_2023, 'happiness_score')
plt.title('Residuals Distribution for 2023')

plt.show()
````

# Part 4: Multiple Regression (16 points)

In this part, we'll use multiple features to do predictions.

&#9989;  **Question 4.1 (5 points)**: Using the `OLS` method in `statsmodels`, perform a multivariable linear regression to predict the `happiness_score` based on `gdp_per_capita`, `healthy_life_expectancy`, `freedom_to_make_life_choices`, and `generosity`. Also, use the `add_constant()` method in `statsmodels` to ensure the model includes a constant term as well. Fit models to both the 2022 and 2023 data and display summaries of both results.

````python
# Put your code here

# Define the independent variables for multivariable regression
independent_vars = ['gdp_per_capita', 'healthy_life_expectancy', 'freedom_to_make_life_choices', 'generosity']

# Add a constant term to the independent variables
X_2022 = sm.add_constant(df_2022[independent_vars])
X_2023 = sm.add_constant(df_2023[independent_vars])

# Fit multivariable linear regression models for 2022 and 2023
model_2022 = sm.OLS(df_2022['happiness_score'], X_2022).fit()
model_2023 = sm.OLS(df_2023['happiness_score'], X_2023).fit()

# Display summaries of both results
print("Summary of Multivariable Linear Regression Model for 2022:")
print(model_2022.summary())
print("\nSummary of Multivariable Linear Regression Model for 2023:")
print(model_2023.summary())
````

&#9989;  **Question 4.3 (2 points)**: Perform the multivariable linear regression for 2023, but this time with only the **most significant features** (p-values less than 0.05). Display the summary of these results.

````python
# Put your code here
import statsmodels.api as sm

# Selecting only the significant features from the previous model for 2023
significant_features_2023 = df_2023[['gdp_per_capita', 'healthy_life_expectancy', 'freedom_to_make_life_choices']]

# Adding a constant term
X_2023_significant = sm.add_constant(significant_features_2023)

# Fitting the multivariable linear regression model
model_2023_significant = sm.OLS(df_2023['happiness_score'], X_2023_significant).fit()

# Displaying the summary of the results
print(model_2023_significant.summary())
````
